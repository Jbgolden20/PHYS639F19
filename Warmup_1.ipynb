The way this one works is by summing up all the nubmers factorable by three and five. It accounts for the duplication by subtracting off one set of factorable by 15

```python
a = 3
b = 5
rangelower = 0
rangeupper = 1000
sum = 0
for i in range (rangelower,rangeupper, 1):
   if i%a == 0:
        sum = sum + i
sumc = 0
for c in range (rangelower,rangeupper, 1):
   if c%b == 0:
        sumc = sumc + c
sumt = 0 
for x in range (rangelower,rangeupper,1):
    if x%(a*b) == 0:
        sumt = sumt + x
print(sum+sumc-sumt)
```

This works by creating an array that is gradualy filled with the fibbati numbers. It usese the last two of the array to make the next item in the array. It then filters by even and less than the peak number (ended up with an extra term at the end) and sums the even in the array

```python
import numpy as np
max = 4000000
fib = np.array([1,1]) #starting Fib numbers
while fib[-1] < max: #only continues when below max
    nextfib = fib[-1]+fib[-2]
    fibnew = np.array([nextfib])
    fib = np.concatenate((fib,fibnew)) #expands fib
fib = fib[ (fib<=max)&(fib%2==0)] #ended up with an extra term, so just set to ingore it
x = np.sum(fib)
print(x)
```

This is my zero finder. It finds the zeros by taking the product of adjecent points and sees if the number is zero or negative. it also reports the total number of zeros

```python
import numpy as np
step=.001
start= -2*np.pi #this is start of test regoion
stop= 1 #this is end of test regoion
i=start
zeros = 0
def function(n):
    return np.sin(n)
# this section is looking for a zero by testing if the product of two points are zero or less
while i<=stop:
    i = i+step
    j = i
    k = i-step
    fx = function(j)*function(k)
    if fx<=0:
        print("this zero is between",round(i-step,3),"and" ,round(i,3))
        zeros=zeros+1 #stores number of intercepts
print("no of zeros is",zeros)
        
```

This works by integrating many small rectangeles. 

```python
import numpy as np
# this defines the function to be intergated
def function(x):
    return np.sin(x)
step = .0001 #this is the incremnt size
lower = 0 #lower bound of intergal
upper = np.pi #upper bound of intergal
i = lower # i is dummy variable
height = 0 #height is the y value that is summed up 
while i < upper-step:
    height = height + function(i)
    i = i + step
print(step*height)
```

This is a very simple global minima finder. Looks for values lower than the stored value. If the value is lower, it stores that value instead

```python
import numpy as np
def function(x): # defines the function to test over
    return np.cos(x)

start = -1 #start point
end = 1 # end point
step= .0001
i = start
fx = function(i) # defines start values
while i <= end:
    if fx >= function(i): #tests to see if new point is smaller or not
        fx = function(i)
        location = i
    i = i + step
print ("Minimum of %s and is at %d" % (fx, location) )
```
